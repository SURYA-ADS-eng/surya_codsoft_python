```python
#task 1: To-Do List application


tasks=[]
while True:
  print("\nTO-DO LIST MENU")
  print("1.Add tasks")
  print("2.View tasks")
  print("3.Delete task")
  print("4.Exit")

  c=int(input("Enter your choice: "))
  if c==1:
    task=input("Enter a task: ")
    tasks.append(task)
    print("Task added successfully")
  elif(c==2):
    if not tasks:
      print("No tasks found")
    else:
      print("your tasks: ")
      i=1
      for task in tasks:
        print(i,".",task)
        i+=1
  elif(c==3):
    if not tasks:
      print("No tasks found to delete")
    else:
      print("Tasks are:")
      i=1
      for task in tasks:
        print(i,".",task)
        i+=1
      num=int(input("Enter the task number to delete: "))
      if 1 <= num <= len(tasks):
        tasks.pop(num-1)
        print("Task deleted successfully")
  elif(c==4):
    print("Exit the program")
    break
  else:
    print("Invalid choice")
    break
```

```python
#task 2- Calculator

n1=float(input("Enter the first number: "))
n2=float(input("Enter the second number: "))
print("Choose any one operation")
print("1.Addition")
print("2.Subtraction")
print("3.Multiplication")
print("4.Division")
c=int(input("Enter your choice (1,2,3,4): "))
if c==1:
         print("Addition is: ",n1+n2)
elif c==2:
         print("Subtraction is: ",n1-n2)
elif c==3:
         print("Multiplication :",n1*n2)
elif c==4:
         if n2==0:
           print("Division by zero is not possible")
         else:
           print("Division is: ",n1/n2)
else:
         print("Invalid choice")
```

```python
#task 3- Password generator

import random
import string
print("PASSWORD GENERATOR")
l=int(input("Enter the length of password: "))
ch=string.ascii_letters+string.digits+string.punctuation
password= ""
for i in range(l):
  password+=random.choice(ch)
print("Your password is: ",password)
```

```python
#task 4- Rock paper scissors game

import random
c=["rock","paper","scissors"]
while True:
  user=input("Enter your choice (rock,paper,scissors): ").lower()
  if user not in c:
    print("Invalid choice")
    continue

  computer=random.choice(c)
  print("computer choose:",computer)
  if user==computer:
    print("It will a tie")
  elif(user=="rock" and computer=="scissors") or (user=="paper" and computer=="rock") or (user=="scissors" and computer=="paper"):
    print("You win the game")
  else:
    print("Computer wins")

  play_again=input("Play again? (yes/no): ").lower()
  if play_again!="yes":
    print("Thanks for playing")
    break
```

```python
#task 5- Contact book

contacts={}
while True:
  print("1.ADD CONTACT")
  print("2.VIEW CONTACTS")
  print("3.SEARCH CONTACT")
  print("4.DELETE CONTACT")
  print("5.EXIT")
  c=int(input("Enter your choice: "))
  if c==1:
    name=input("Enter the name: ").lower()
    phone=input("Enter the phone number: ")
    email=input("Enter the gmail: ")
    contacts[name]={"phone":phone,"email":email}
    print("Contact added successfully")
  elif c==2:
    for name in contacts.items():
      print("Name:",name)
      #print("Phone:",phone)
      #print("Email:",email)
  elif c==3:
    name=input("Enter the name to search: ")
    if name in contacts:
      print(contacts[name])
    else:
      print("Contact not found")
  elif c==4:
    name=input("Enter the name to delete: ")
    contacts.pop(name,None)
    print("Contact deleted successfully")
  elif c==5:
    print("Exit the program")
    break
```
